```python
import pandas as pd
# import glob
# import numpy as np
# from scipy.stats import shapiro
# from scipy.stats import uniform
# import seaborn as sb
# import matplotlib.pyplot as plt
import plotly.figure_factory as ff

df3 = pd.read_excel('D:\Projects_for_my_science_work\Article_1_Clustering\data_of_community\data_vk_2020_2021\Vk_data_all_type_test_1\df_main_feat_goal.xlsx', header=1, index_col=0)
df3_c = df3.copy()
df3_c = df3_c.drop_duplicates(subset=['Cообщества'], ignore_index=True)

df3_c["Тематика"] = df3_c["Тематика"].astype('category')
df3_c["Cообщества"] = df3_c["Cообщества"].astype('category')

df3_c["Тематика_id"] = df3_c["Тематика"].cat.codes
df3_c["Cообщества_id"] = df3_c["Cообщества"].cat.codes
df3_c = df3_c.drop(columns = ['Cообщества','Тематика','Есть верификация?'],axis = 1)
```

```python
colorscales = [
    "Greys",
    "YlGnBu",
    "Greens",
    "YlOrRd",
    "Bluered",
    "RdBu",
    "Reds",
    "Blues",
    "Picnic",
    "Rainbow",
    "Portland",
    "Jet",
    "Hot",
    "Blackbody",
    "Earth",
    "Electric",
    "Viridis",
    "Cividis",
]
```

```python
from plotly.offline import iplot

corrs = df3_c.corr()
# print(corrs)
figure = ff.create_annotated_heatmap(
    z=corrs.values,
    x=list(corrs.columns),
    y=list(corrs.index),
    colorscale="Earth",
    annotation_text=corrs.round(2).values,
    showscale=True,
    reversescale=True,
)
figure.layout.margin = dict(l=200, t=200)
figure.layout.height = 800
figure.layout.width = 1000

iplot(figure)
```
