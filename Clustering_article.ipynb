##  1. Объединение файлов по 200+ сообществ в один общий фрейм

```python
import pandas as pd
import glob
import numpy as np
import os.path
import random
from collections import defaultdict
from datetime import datetime
import time
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler

def record_excel(df,str):
    writer = pd.ExcelWriter(str + '.xlsx', engine='xlsxwriter')
    df.to_excel(writer, sheet_name='Sheet2')
    writer.save()

df_main=pd.DataFrame();df_attached=pd.DataFrame()

start_time = datetime.now()
path = 'D:\Projects_for_my_science_work\Article_1_Clustering\data'
num_files = len([f for f in os.listdir(path)
                if os.path.isfile(os.path.join(path, f))])
print('Files of directory:',num_files)
print('\n\n')

count_file = 0
for f in glob.glob('D:\Projects_for_my_science_work\Article_1_Clustering\data\*.xlsx'):
    #print(f)
    print('\n\n')
    count_file = count_file+1
    df = pd.read_excel(f, index_col=None, header=None, usecols=None)
    print('1. Number of current file of this directory:',count_file)
    print('2. Count сommunities of current file: ',len(df.index))
    if count_file > 1:
        df_attached = df.iloc[ 2 : len(df.index) , ]
        df_main = df_main.append(df_attached, ignore_index=True)
    else:
        df_main = df_main.append(df, ignore_index=True)
df_main = df_main.drop(labels = [0],axis = 0)
df_main = df_main.drop(labels = [0],axis = 1)

record_excel( df_main, 'df_main' )
```

##  2. Формирование фрейма данных, приведение порядковых, номинальных и бинарных признаков к количественным
##### * на выходе получается фрейм готовый к применению

```python
data_main = pd.read_excel('df_main.xlsx', header=1, index_col=0)
data = data_main.copy()
record_excel( data, 'data' )
cols = pd.DataFrame(data.columns)
record_excel( cols, 'cols' )
```

```python
data_c = data.copy()
```

```python
data_c = data_c.drop_duplicates(subset=['Cообщества'], ignore_index=True)

data_c["Тематика"] = data_c["Тематика"].astype('category')
data_c["Cообщества"] = data_c["Cообщества"].astype('category')
data_c["Период"] = data_c["Период"].astype('category')
data_c["URL"] = data_c["URL"].astype('category')

data_c["Тематика_id"] = data_c["Тематика"].cat.codes
data_c["Cообщества_id"] = data_c["Cообщества"].cat.codes
data_c["Период_id"] = data_c["Период"].cat.codes
data_c["URL_id"] = data_c["URL"].cat.codes

record_excel( data_c, 'data_c' )
```

```python
data_c = data_c.drop(columns = ['Cообщества','Тематика','Период','URL', 'Есть верификация?'],axis = 1)
record_excel( data_c, 'data_c' )
feat_name = pd.DataFrame(data_c.columns)
record_excel( pd.DataFrame(feat_name), 'feat_name' )
```

```python
data_c = data_c.replace({np.nan: None})
data_c = data_c.fillna(0)
data_c = data_c.sort_values(by='Тематика_id')
record_excel( data_c, 'data_converted' )
```
